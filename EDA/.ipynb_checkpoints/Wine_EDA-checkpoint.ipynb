## Cleaning the Dataset 

```python
with open("Data/winequality_red.csv", "r", encoding="utf-8-sig") as f: 
    text = f.read()
    
clean_text = text.replace('"', "")
wine_df = pd.read_csv(StringIO(clean_text), sep=';')
wine_df.head()
```

```python
# Checking for null rows
print("Shape:", wine_df.shape)
print("\nColumns:")
print(wine_df.columns.tolist())

null_counts = wine_df.isnull().sum()

print("Null values per column:")
print(null_counts)

print("\nTotal null values:", null_counts.sum())
```

```python
# Ensure Data directory exists
os.makedirs("Data", exist_ok=True)

# Save cleaned dataframe
output_path = "Data/wine_df.csv"
wine_df.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to: {output_path}")
```

## Target Variable Exploration

```python
# raw distribution of target variable
plt.figure(figsize=(6, 4))
wine_df["quality"].value_counts().sort_index().plot(kind="bar")
plt.xlabel("Wine Quality Score")
plt.ylabel("Count")
plt.title("Distribution of Original Wine Quality Scores")
plt.tight_layout()
plt.show()
```

```python
# 3-class Labeled Distribution
def make_3class_labels(y_quality):
    """
    Convert original wine quality scores into 3 classes:
      0 = low (<=5)
      1 = medium (=6)
      2 = high (>=7)
    """
    yq = pd.Series(y_quality).astype(int).to_numpy()

    y = np.where(
        yq <= 5, 0,
        np.where(yq == 6, 1, 2)
    )

    class_names = ["low (<=5)", "mid (=6)", "high (>=7)"]
    return y, class_names

y, class_names = make_3class_labels(wine_df["quality"])
X = wine_df.drop(columns=["quality"])
pd.Series(y).value_counts().sort_index()
```

```python
## 3-class Label Distribution
plt.figure(figsize=(5, 4))
pd.Series(y).value_counts().sort_index().plot(kind="bar")
plt.xticks(range(3), class_names, rotation=0)
plt.ylabel("Count")
plt.title("3-Class Wine Quality Distribution")
plt.tight_layout()
plt.show()
```

## Feature Distribution Analysis

```python
# All feature columns
features = X.columns
n_features = len(features)

# Choose grid size automatically
n_cols = 4
n_rows = math.ceil(n_features / n_cols)

plt.figure(figsize=(4 * n_cols, 3 * n_rows))

for i, col in enumerate(features, 1):
    plt.subplot(n_rows, n_cols, i)
    plt.hist(X[col], bins=20)
    plt.title(col)
    plt.xlabel("")
    plt.ylabel("")

plt.tight_layout()
plt.show()
```

```python
features = list(X.columns)
features_per_fig = 1

for start in range(0, len(features), features_per_fig):
    subset = features[start:start + features_per_fig]

    plt.figure(figsize=(10, 4))

    for i, col in enumerate(subset, 1):
        plt.subplot(1, features_per_fig, i)
        plt.boxplot(
            [X[col][y == c] for c in range(3)],
            labels=class_names
        )
        plt.title(col)
        plt.xticks(rotation=0)

    plt.tight_layout()
    plt.show()


```

```python
corr = X.corr().values
cols = X.columns
n = len(cols)

mask = np.triu(np.ones_like(corr, dtype=bool), k=1)
corr_masked = np.ma.array(corr, mask=mask)

plt.figure(figsize=(10, 8))
plt.imshow(corr_masked, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(label='Correlation')

plt.xticks(range(n), cols, rotation=90)
plt.yticks(range(n), cols)
plt.title("Lower-Triangle Feature Correlation Matrix")

for i in range(n):
    for j in range(i + 1):
        plt.text(j, i, f"{corr[i, j]:.2f}", ha="center", va="center", color="black", fontsize=8)

plt.tight_layout()
plt.show()

```
